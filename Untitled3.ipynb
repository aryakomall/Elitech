```python
import pandas as pd

file_path = 'final_combined_data (2).xlsx'
df = pd.read_excel(file_path)
```

```python
categories = ['TYRE', 'TYRE REPAIR RUBBER', 'TYRE REPAIR', 'TUBE', 'AIR COMPRESSOR']
filtered_df = df[df['CATOGARY OF ITEM'].isin(categories)]
```

```python
filtered_df['ITEM NAME'].value_counts()
```

```python
# prompt: give the monthly analysis for consumption and bought items

monthly_consumption = filtered_df.groupby('DATE')['QTY (OUTWARD)'].sum().resample('M').sum()
monthly_bought = filtered_df.groupby('DATE')['QTY (INWARD)'].sum().resample('M').sum()

# Create a new DataFrame for monthly analysis
monthly_analysis_df = pd.DataFrame({
    'Month': monthly_consumption.index,
    'Consumption': monthly_consumption.values,
    'Bought': monthly_bought.values
})

# Print the monthly analysis DataFrame
print(monthly_analysis_df)

# Plot the monthly consumption and bought trends
plt.figure(figsize=(10, 6))
sns.lineplot(data=monthly_analysis_df, x='Month', y='Consumption', label='Consumption', marker='o')
sns.lineplot(data=monthly_analysis_df, x='Month', y='Bought', label='Bought', marker='s')

# Add annotations
for x, y in zip(monthly_analysis_df['Month'], monthly_analysis_df['Consumption']):
    plt.text(x, y, f'{y:.0f}', fontsize=9, ha='right')

for x, y in zip(monthly_analysis_df['Month'], monthly_analysis_df['Bought']):
    plt.text(x, y, f'{y:.0f}', fontsize=9, ha='right')

plt.title('Monthly Consumption and Bought Items Analysis')
plt.xlabel('Month')
plt.ylabel('Total Quantity')
plt.legend()
plt.show()
```

```python
df_selected = filtered_df.dropna(subset=['QTY (OUTWARD)'])
df_selected = df_selected[['CATOGARY OF ITEM', 'QTY (OUTWARD)', 'DATE']]

# Convert DATE column to datetime if it's not already
df_selected['DATE'] = pd.to_datetime(df_selected['DATE'], errors='coerce')

# Group by ITEM NAME to calculate total quantities used
grouped_df = df_selected.groupby('CATOGARY OF ITEM').agg({
    'QTY (OUTWARD)': 'sum'
}).reset_index()

# Find top 10 most used items
top_used_items = grouped_df.sort_values(by='QTY (OUTWARD)', ascending=False).head(10)

# Separate plot for each top used item
plt.figure(figsize=(16, 12))

for i, item in enumerate(top_used_items['CATOGARY OF ITEM']):
    plt.subplot(5, 2, i + 1)
    item_data = df_selected[df_selected['CATOGARY OF ITEM'] == item]
    item_data.set_index('DATE', inplace=True)
    monthly_data = item_data.resample('M').sum()

    sns.lineplot(data=monthly_data['QTY (OUTWARD)'], marker='o', label=item)
    plt.title(f'Monthly Usage Trend for {item}')
    plt.xlabel('Date')
    plt.ylabel('Total Quantity Used')
    plt.legend()

plt.tight_layout()
plt.show()
```
